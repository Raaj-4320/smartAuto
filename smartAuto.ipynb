### There are lot of similarities in the names of drivetrain, for example Front-wheel Drive, FWD, Front-Wheel Drive, Front Wheel Drive is a same thing.

 We can fix this by the use of regex.

```python
df['drivetrain'].value_counts()
```

```python
# Define a mapping for the changes
mapping = {
    # Set one of change
    'Front-wheel Drive': 'Front Wheel Drive',
    'FWD': 'Front Wheel Drive',
    'Front-Wheel Drive': 'Front Wheel Drive',
    'Front Wheel Drive': 'Front Wheel Drive',
    '4x2': 'Front Wheel Drive',
    '2WD': 'Front Wheel Drive',

    # Set two of change
    'All-wheel Drive': 'All Wheel Drive',
    'AWD': 'All Wheel Drive',
    'All-Wheel Drive': 'All Wheel Drive',
    'All Wheel Drive': 'All Wheel Drive',

    # Set three of change
    'Four-wheel Drive': 'Four Wheel Drive',
    'Four-Wheel Drive': 'Four Wheel Drive',
    '4WD': 'Four Wheel Drive',
    'Four Wheel Drive': 'Four Wheel Drive',
    '4x4': 'Four Wheel Drive',
    '4X4': 'Four Wheel Drive',

    # Set four of change
    'Rear-wheel Drive': 'Rear Wheel Drive',
    'RWD': 'Rear Wheel Drive',
    'Rear-Wheel Drive': 'Rear Wheel Drive',
    'Rear Wheel Drive': 'Rear Wheel Drive'
}

# Replace values based on mapping, keep others unchanged
df['standardized_drivetrain'] = df['drivetrain'].replace(mapping)
```

```python
df['standardized_drivetrain'].value_counts()
```

### So far we were able to shrink the number of unique values of differnt drivetrain which were ultimately making same meaning.
